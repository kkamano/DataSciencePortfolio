# Basic analysis and visualization using Animal Crossing Datasets

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', 100)
```

## Read data

```python
%ls data
```

```python
# read villagers data
df_vil = pd.read_csv('./data/villagers.csv')
df_vil = df_vil.drop(columns=['Furniture List','Filename','Unique Entry ID'])
display(df_vil.head())
print(df_vil.shape)
print(df_vil.describe())
```

```python
# read villagers' popularity data
df_pop = pd.read_csv('./data/acnh_villager_data.csv')
display(df_pop.head())
print(df_pop.shape)
print(df_pop.describe())
```

- Need to add column 'overall rank'. (It has 413 records but maximum rank is 163.)

```python
df_pop[:16]
```

- looks like it has ranking in tier

```python
df_pop = df_pop.sort_values(by=['tier','rank']).reset_index()
df_pop = df_pop.rename(columns={'index':'overall_rank'})
df_pop['overall_rank'] = df_pop['overall_rank'] + 1

df_pop = df_pop.sort_values(by=['overall_rank'],ascending=False).reset_index(drop=True)
df_pop = df_pop.reset_index()
df_pop = df_pop.rename(columns={'index':'popularity'})
df_pop['popularity'] = df_pop['popularity'] + 1

display(df_pop.head())
print(df_pop.shape)
print(df_pop.describe())
```

```python
# merge 2 dataframe
df_pop = df_pop.rename(columns={'name':'Name'})

df = pd.merge(df_vil,df_pop,on=['Name'],how='left')
df = df.sort_values(by=['rank']).reset_index(drop=True)
display(df.head())
print(df.shape)
```

## Visualizing data

```python
df
```

```python
df.columns
```

### Countplot of some columns

```python
target_cols = ['Species', 'Gender', 'Personality', 'Hobby', 
              'Style 1', 'Style 2', 'Color 1','Color 2']
for col in target_cols:
    plt.figure(figsize=(15,5))
    ax = sns.countplot(x=col,data=df)
    plt.xticks(rotation=45)
    plt.show()
```

### boxplot

```python
target_cols = ['Species', 'Gender', 'Personality', 'Hobby', 
              'Style 1', 'Style 2', 'Color 1','Color 2']
for col in target_cols:
    plt.figure(figsize=(15,5))
    sns.boxplot(x=col, y="popularity", data=df)
    plt.xticks(rotation=45)
    plt.show()
```

### Scatter plot

```python
target_cols = ['Species', 'Gender', 'Personality', 'Hobby', 
              'Style 1', 'Style 2', 'Color 1','Color 2']
for col in target_cols:
    tmp = df.copy()
    tmp['size'] = 1
    tmp = tmp.groupby(by=col,as_index=False).agg({'popularity':'mean','size':'sum'})
    tmp = tmp.sort_values(by='popularity',ascending=False)
    plt.figure(figsize=(15,5))
    sns.scatterplot(x=col, y="popularity", size='size',data=tmp, sizes=(100, 300))
    plt.xticks(rotation=45)
    plt.show()
```

## Summary

- Cute animals have more popularity. (Species, Style1, Style2)
